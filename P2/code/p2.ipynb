```python
import matplotlib.pyplot as plt
from networkx import nx
import random
import matplotlib.pyplot as plt
import os
import collections
import numpy as np
from scipy.special import factorial
```

```python
def draw_graph_node_size(G,x,y):
    pos = nx.spring_layout(G, iterations=200)
    node_sizes =[10 + 20 * i for i in range(len(G))]
    M = G.number_of_edges()
    edge_colors = range(2, M + 2)
    edge_alphas = [(5 + i) / (M + 4) for i in range(M)]
    d = dict(G.degree)


    plt.figure(figsize=(x,y))
    nx.draw(G, nodelist=d.keys(), node_size=[v * 100 for v in d.values()], node_color='purple', alpha=0.7, edge_color='green')
    #nodes = nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='purple', alpha=0.7)
    #edges = nx.draw_networkx_edges(G, pos, node_size=node_sizes, edge_color='green', width=2, edge_alphas=0.9)

    ax = plt.gca()
    ax.set_axis_off()
    plt.show()
    
```

```python

#Path for all graphs generated, with this notebook and the network.py cli
path = '../graphs/'
files = []
# r=root, d=directories, f = files
index = 0
for r, d, f in os.walk(path):
    for file in f:
        if '.csv' in file:
            files.append(os.path.join(r, file))
            print('{:02d}'.format(index), file) #prints the folder and the file is contained
            index = index + 1
```

```python
g_list = []
graph_map = {}

ba_dict = {}      #0:3
steps_500_3 = {}  #50:71
steps_500_4 = {}  #72:93
steps_5000_3 = {} #4:26
steps_5000_4 = {} #27:49
er_dict = {}#     94:

for i in range(len(files)):
    g_list.append(nx.read_multiline_adjlist(files[i],delimiter=','))
    graph_map[files[i]] = g_list[i]
    
ba_list      = g_list[:94]
er_list      = g_list[94:]

index = 0
for k, v in graph_map.items():
    if index in range(0,4):
        ba_dict[k] = v
    elif index in range(50,72):
        steps_500_3[k] = v
    elif index in range(72,94):
        steps_500_4[k] = v
    elif index in range(4,27):
        steps_5000_3[k] = v
    elif index in range(27,50):
        steps_5000_4[k] = v
    else:
        er_dict[k] = v
    index = index+1
```

```python
avg = []
index = 0
total = 5

N = 0
L = 0
k = 0
D = 0
N_c = 0
LHS = 0
d = 0
C = 0
N_g = 0

for f,g in er_dict.items():
    
    #Metricas
    degree_sequence = sorted([d for n, d in g.degree()], reverse=True)  # degree sequence
    degreeCount = collections.Counter(degree_sequence)
    deg, cnt = zip(*degreeCount.items())
    N += nx.number_of_nodes(g)
    L += nx.number_of_edges(g)
    k += np.mean(degree_sequence)
    D += nx.density(g)
    N_c += nx.number_connected_components(g)
    LHS += max(degreeCount)
    try:
        d += nx.average_shortest_path_length(g)
    except:
        d = 0
    C += np.mean(degree_sequence) / N
    N_g += max(nx.connected_component_subgraphs(g), key=len).number_of_nodes()

    if index % 5 == 4: #Si llegamos al ultimo hacemos la media e imprimimos el ultimo
        print(f.split('\\')[1].split('/')[0])
        print("----------Teorico----------")
        Nt = int(f.split('\\')[2].split('_')[3][1:])
        P = float("0.{}".format(f[-11:-4].split('.')[1]))
        Lt = P * ((Nt*(Nt-1)/2))
        print("N:",Nt)
        print("P:",P)
        print("<L>:",Lt)
        print("<k>:",P*(Nt-1))
        print("D:", Lt/((Nt*(Nt-1)/2)))
        print("N_c:", N_c/total)
        #print("Largest Hub Size:", LHS/total)
        if d != 0:
            print("<d>:",np.log(Nt)/np.log(P*(Nt-1)))
        else:
            print("<d>: -")
        print("C:",P*(Nt-1) / Nt)
       
        #print("N_g:", N_g/total)
        print("---------------------------")
        print("N:",N/total)
        print("<L>:", L/total)
        print("<k>:",k/total)
        print("D:", D/total)
        print("N_c:", N_c/total)
        #print("Largest Hub Size:", LHS/total)
        if d != 0:
            print("<d>:",d/total)
        else:
            print("<d>: -")
        print("C:",C/total)
       
        #print("N_g:", N_g/total)
        print("---------------------------")
        N = 0
        L = 0
        k = 0
        D = 0
        N_c = 0
        LHS = 0
        d = 0
        C = 0
        N_g = 0
        fig, ax = plt.subplots()
        plt.bar(deg, cnt, width=0.80, color='b')
        plt.plot(deg, cnt, marker='o',linestyle='dashed', color='purple',markersize=5, alpha=0.6)

        plt.title("Degree Distribution")
        plt.ylabel("Count")
        plt.xlabel("Degree")
        #ax.set_xticks([d for d in deg])
        #ax.set_xticklabels(deg)
        
    index += 1
    if False:
        # draw graph in inset
        plt.axes([0.4, 0.4, 0.5, 0.5])
        Gcc = g.subgraph(sorted(nx.connected_components(g), key=len, reverse=True)[0])
        pos = nx.spring_layout(g)
        plt.axis('off')
        nx.draw_networkx_nodes(g, pos, node_size=20, node_color='purple', alpha=0.5)
        nx.draw_networkx_edges(g, pos, alpha=0.4, edge_color='green')

    plt.show()
```
